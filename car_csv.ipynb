```python
sns.pairplot(car)
```

```python
car=car.drop(['Fuel Information.Fuel Type','Fuel Information.Fuel Type'],axis=1)
car.head(5)
```

```python
car=car.rename(columns={"Identification.ID":"ID","Dimensions.Height":"Height"})
car.head(5)
```

```python
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
car.count()
```

```python
car=car.drop_duplicates()
car.head()
```

```python
print(car.isnull().sum())
```

```python
car=car.dropna()
car.count()
```

```python
car.Height.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("number of car by Height")
plt.ylabel("Number of cars")
plt.xlabel("Height");
```

```python
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
```
